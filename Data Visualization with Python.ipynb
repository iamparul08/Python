```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x = [1, 2, 3]
y = [4, 5, 9]
z = [10, 5, 0]
plt.plot(x, y) #plotting x & y(blue line)
plt.plot(x, z) #plotting x &z (orange line)
plt.title("test plot") #name 
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["this is y", "this is z"]) #this is y & this is z 
plt.show() #show function
```
